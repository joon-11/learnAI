```python
train_df = train_df.drop(['Ticket', 'Cabin'], axis=1)
test_df = test_df.drop(['Ticket', 'Cabin'], axis=1)
combine = [train_df, test_df]
print("After", train_df.shape, test_df.shape)
```

```python
train_df = train_df.drop(['Name', 'PassengerId'], axis=1)
test_df = test_df.drop(['Name'], axis=1)
combine = [train_df, test_df]
train_df.shape, test_df.shape
```

```python
train_df = train_df.fillna({'contact' : 'unknown'})
test_df = test_df.fillna({'contact' : 'unknown'})
train_df.head()
```

```python
train_df = train_df.replace('female', 1)
train_df = train_df.replace('male', 0)

test_df = test_df.replace('female', 1)
test_df = test_df.replace('male', 0)

train_df.head()
```

```python
train_df = train_df.replace('S', 0)
train_df = train_df.replace('C', 1)
train_df = train_df.replace('Q', 2)

test_df = test_df.replace('S', 0)
test_df = test_df.replace('C', 1)
test_df = test_df.replace('Q', 2)
```

```python
def bar_chart(feature):
    survived = train_df[train_df['Survived']==1][feature].value_counts()
    dead = train_df[train_df['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline

bar_chart('Sex')
```

```python
surP = train_df[train_df['Survived'] == 1]
surM = surP[train_df['Sex'] == 0].count()[0]
surG = surP[train_df['Sex'] == 1].count()[0]
Sum = train_df.count()
# print(surP.count()[0])
print(surM/surP.count()[0]*100)
print(surG/surP.count()[0]*100)
```

```python
X_train = train_df.drop("Survived", axis=1)
Y_train = train_df["Survived"]
X_test  = test_df.drop("PassengerId", axis=1).copy()
X_train.shape, Y_train.shape, X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
acc_decision_tree
```

```python
submission = pd.DataFrame({
        "PassengerId": test_df["PassengerId"],
        "Survived": Y_pred
    })

submission.to_csv('submision', index = False)
```
